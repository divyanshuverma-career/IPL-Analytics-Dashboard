# Step 2: Making multiple table out of raw dataset

```python
ball_by_ball = df.copy()
```

```python
drop_cols = [

    'review_batter',
    'team_reviewed',
    'review_decision',
    'umpire',
    'umpires_call',
    'power_surge_start',
    'new_batter',
    'next_batter'
]

ball_by_ball.drop(columns=drop_cols, inplace=True)
```

```python
team_mapping = {
    'Royal Challengers Bangalore': 'Royal Challengers Bengaluru'

}

team_cols = [
    'batting_team',
    'bowling_team',
    'toss_winner',
    'match_won_by'
]

for col in team_cols:
    ball_by_ball[col] = (
        ball_by_ball[col]
        .replace(team_mapping)
    )
```

```python
venue_mapping = {
    'Maharaja Yadavindra Singh International Cricket Stadium, Mullanpur':
    'Maharaja Yadavindra Singh International Cricket Stadium, New Chandigarh'
}

ball_by_ball['venue'] = (
    ball_by_ball['venue']
    .replace(venue_mapping)
)
```

```python
ball_by_ball['date'] = pd.to_datetime(
    ball_by_ball['date']
)
```

```python
def get_phase(over):

    if over <= 5:
        return 'Powerplay'

    elif over <= 14:
        return 'Middle Overs'

    else:
        return 'Death Overs'

ball_by_ball['phase'] = (
    ball_by_ball['over']
    .apply(get_phase)
)
```

```python
ball_by_ball['display_over'] = (
    ball_by_ball['over'] + 1
)
```

```python
ball_by_ball['is_boundary'] = (
    ball_by_ball['runs_batter']
    .isin([4, 6])
)
```

```python
ball_by_ball['is_dot_ball'] = (
    ball_by_ball['runs_total'] == 0
)
```

```python
# Create Legal Wicket Column
# Bowler shold not get wicket for run_out or retired_hurt



valid_bowler_wickets = [
    'bowled',
    'caught',
    'lbw',
    'stumped',
    'caught and bowled',
    'hit wicket'
]

ball_by_ball['is_bowler_wicket'] = (
    ball_by_ball['wicket_kind']
    .isin(valid_bowler_wickets)
)
```

```python
ball_by_ball.to_csv(
    'ball_by_ball.csv',
    index=False
)
```

# Table 2- match_summary

```python
match_summary = (
    ball_by_ball[
        [
            'match_id',
            'date',
            'season',
            'venue',
            'city',
            'stage',
            'toss_winner',
            'toss_decision',
            'match_won_by',
            'player_of_match',
            'win_outcome',
            'result_type',
            'method'
        ]
    ]
    .drop_duplicates()
)
```

```python
teams = (
    ball_by_ball.groupby('match_id')['batting_team']
    .unique()
    .reset_index()
)

teams['team_count'] = teams['batting_team'].apply(len)
```

```python
teams = teams[teams['team_count'] == 2]
```

```python
teams['team1'] = teams['batting_team'].apply(
    lambda x: x[0]
)

teams['team2'] = teams['batting_team'].apply(
    lambda x: x[1]
)
```

```python
teams = teams[['match_id', 'team1', 'team2']]

match_summary = match_summary.merge(
    teams,
    on='match_id',
    how='inner'
)
```

```python
match_summary['match_id'].nunique()
```

Creating inning score table:

```python
# it captures total runs scored in each innings

innings_score = (
    ball_by_ball.groupby(
        ['match_id', 'innings']
    )['runs_total']
    .sum()
    .reset_index()
)
```

```python

# Pivot converts row values --> columns


innings_pivot = (
    innings_score.pivot(
        index='match_id',
        columns='innings',
        values='runs_total'
    )
    .reset_index()
)
```

```python
innings_pivot.columns  # here 3, 4 are the super over innings
```

```python
innings_pivot = innings_pivot[
    ['match_id', 1, 2]
]
```

```python
innings_pivot.columns = [
    'match_id',
    'innings1_score',
    'innings2_score'
]
```

```python
# Merge scorres - ignoring super oper for now

match_summary = match_summary.merge(
    innings_pivot,
    on='match_id',
    how='left'
)
```

```python
innings_pivot.columns
```

```python
match_summary.columns
```

```python
# chase vs defend
import numpy as np

match_summary['win_type'] = np.where(
    match_summary['innings2_score'] >
    match_summary['innings1_score'],

    'Chased',

    'Defended'
)
```

```python
# toss advantage

match_summary['toss_match_same_winner'] = np.where(
    match_summary['toss_winner'] ==
    match_summary['match_won_by'],

    'Yes',

    'No'
)
```

# Table 3 : Batter_stat

```python
batter_stat = (
    ball_by_ball.groupby(
        [
            'match_id',
            'season',
            'batting_team',
            'batter'
        ]
    )
    .agg(
        runs=('runs_batter', 'sum'),

        balls=('valid_ball', 'sum'),

        fours=('runs_batter',
               lambda x: (x == 4).sum()),

        sixes=('runs_batter',
               lambda x: (x == 6).sum()),

        dismissals=('player_out',
                     lambda x: x.notnull().sum())
    )
    .reset_index()
)
```

```python
# Strike rate

batter_stat['strike_rate'] = (
    batter_stat['runs']
    / batter_stat['balls']
) * 100
```

```python
# Boundary runs

batter_stat['boundary_runs'] = (
    batter_stat['fours'] * 4
    +
    batter_stat['sixes'] * 6
)
```

```python
#boundary percentage
batter_stat['boundary_pct'] = np.where(
    batter_stat['runs'] > 0,

    (batter_stat['boundary_runs']
     / batter_stat['runs']) * 100,

    0
)
```

```python
batter_stat['fifty'] = (
    batter_stat['runs'] >= 50
)
```

```python
batter_stat['century'] = (
    batter_stat['runs'] >= 100
)
```

```python
batter_stat.info()
```

```python
batter_stat.sort_values(
    by='runs',
    ascending=False
).head(10)
```

```python
batter_stat.to_csv(
    'batter_stat.csv',
    index=False
)
```

# Bowler_stat

```python
bowler_stat = (
    ball_by_ball.groupby(
        [
            'match_id',
            'season',
            'bowling_team',
            'bowler'
        ]
    )
    .agg(
        balls=('valid_ball', 'sum'),

        runs_conceded=('runs_total', 'sum'),

        wickets=('is_bowler_wicket', 'sum'),

        dot_balls=('is_dot_ball', 'sum')
    )
    .reset_index()
)
```

```python
# Over bowled

bowler_stat['overs'] = (
    bowler_stat['balls'] / 6
)
```

```python
bowler_stat['economy'] = (
    bowler_stat['runs_conceded']
    / bowler_stat['overs']
)
```

```python
bowler_stat['economy'] = np.where(
    bowler_stat['overs'] > 0,

    bowler_stat['runs_conceded']
    / bowler_stat['overs'],

    0
)
```

```python
bowler_stat['dot_ball_pct'] = np.where(
    bowler_stat['balls'] > 0,

    (bowler_stat['dot_balls']
     / bowler_stat['balls']) * 100,

    0
)
```

```python
bowler_stat['bowling_average'] = np.where(
    bowler_stat['wickets'] > 0,

    bowler_stat['runs_conceded']
    / bowler_stat['wickets'],

    0
)
```

```python
bowler_stat['three_wicket_haul'] = (
    bowler_stat['wickets'] >= 3
)
```

```python
bowler_stat['five_wicket_haul'] = (
    bowler_stat['wickets'] >= 5
)
```

```python
bowler_stat.head()
```

```python
bowler_stat.info()
```

```python
bowler_stat.sort_values(
    by='wickets',
    ascending=False
).head(10)
```

```python
bowler_stat.to_csv(
    'bowler_stat.csv',
    index=False
)
```

# Team_performance

```python
team_matches = pd.concat([

    match_summary[['match_id', 'team1']]
    .rename(columns={'team1': 'team'}),

    match_summary[['match_id', 'team2']]
    .rename(columns={'team2': 'team'})

])
```

```python
matches_played = (
    team_matches.groupby('team')
    .agg(
        matches_played=('match_id', 'nunique')
    )
    .reset_index()
)
```

```python
matches_won = (
    match_summary.groupby('match_won_by')
    .agg(
        matches_won=('match_id', 'count')
    )
    .reset_index()
    .rename(columns={'match_won_by': 'team'})
)
```

```python
team_performance = matches_played.merge(
    matches_won,
    on='team',
    how='left'
)
```

```python
team_performance['matches_won'] = (
    team_performance['matches_won']
    .fillna(0)
)
```

```python
team_performance['win_pct'] = (
    team_performance['matches_won']
    / team_performance['matches_played']
) * 100
```

```python
# Here, team run calculating

team_runs = (
    ball_by_ball.groupby('batting_team')
    .agg(
        total_runs=('runs_total', 'sum')
    )
    .reset_index()
    .rename(columns={'batting_team': 'team'})
)
```

```python
team_wickets = (
    ball_by_ball.groupby('bowling_team')
    .agg(
        total_wickets=('is_bowler_wicket', 'sum')
    )
    .reset_index()
    .rename(columns={'bowling_team': 'team'})
)
```

```python
team_fours = (
    ball_by_ball[
        ball_by_ball['runs_batter'] == 4   # chauka
    ]
    .groupby('batting_team')
    .size()
    .reset_index(name='fours')
    .rename(columns={'batting_team': 'team'})
)
```

```python
team_sixes = (
    ball_by_ball[
        ball_by_ball['runs_batter'] == 6    # for chakka
    ]
    .groupby('batting_team')
    .size()
    .reset_index(name='sixes')
    .rename(columns={'batting_team': 'team'})
)
```

```python
team_performance = (
    team_performance
    .merge(team_runs, on='team', how='left')
    .merge(team_wickets, on='team', how='left')
    .merge(team_fours, on='team', how='left')
    .merge(team_sixes, on='team', how='left')
)
```

```python
team_performance = (
    team_performance.fillna(0)
)
```

```python
team_performance.head()
```

```python
team_performance.sort_values(
    by='matches_won',
    ascending=False
)
```

```python
team_performance.to_csv(
    'team_performance.csv',
    index=False
)
```
